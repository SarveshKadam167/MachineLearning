# Model Evaluation

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg_1 = LinearRegression()
lin_reg_1.fit(X_train, y_train)
y_pred_mlr = lin_reg_1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_mlr)
```

## Polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y_train)
y_pred_plr = lin_reg_2.predict(poly_reg.transform(X_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_plr)
```

##Support Vector Regression

```python
from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(len(y_train),1))
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_train_scaled)
```

```python
print(y_train_scaled)
```

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel = 'rbf')
svr_reg.fit(X_train_scaled,y_train_scaled)
X_test_scaled = scaler_X.transform(X_test)
y_test_scaled = scaler_y.transform(y_test.reshape(len(y_test),1))
y_pred_svr = svr_reg.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test_scaled, y_pred_svr)
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dtr_reg = DecisionTreeRegressor(random_state=0)
dtr_reg.fit(X_train,y_train)
y_pred_dtr = dtr_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_dtr)
```

## Support Vector Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr_reg = RandomForestRegressor(n_estimators=10, random_state=0)
rfr_reg.fit(X_train,y_train)
y_pred_rfr = rfr_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_rfr)
```
